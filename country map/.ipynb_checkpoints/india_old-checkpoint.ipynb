## India

```python
import geopandas as gpd
import matplotlib.pyplot as plt

import colorcet as ctx
import contextily as ct
```

```python
gdf = gpd.read_file('../data/india_admin_areas.geojson')
gdf.head(3)
```

```python
gdf['st_nm'].count()
```

```python
states = gdf.dissolve(by='st_nm').reset_index()
states = states.to_crs('EPSG:7755')

states['centroid'] = states.geometry.centroid
states['st_nm'].count()
```

```python
fig_1, ax_1 = plt.subplots(figsize=(6,8))
states.plot(ax=ax_1, column='st_nm', categorical=True, cmap=ctx.cm.glasbey_cool)

for row in states.itertuples():
    x, y = row.centroid.x, row.centroid.y
    label = row.st_nm

    if label in ['Chandigarh']:
        x += 1 * 10**5

    ax_1.annotate(label, xy=(x,y), fontsize=5, ha='center')

plt.title('India: States and Union Territories (EPSG: 7755)')
plt.xlabel('Easting (m)')
plt.ylabel('Northing (m)')

plt.tight_layout()
plt.show()
```

```python
fig_2, ax_2 = plt.subplots(figsize=(6,8))
states.plot(ax=ax_2, column='st_nm', categorical=True, cmap=ctx.cm.glasbey_cool)

basemap = ct.providers.Esri.WorldImagery.copy()
basemap['attribution'] = 'Esri'
ct.add_basemap(ax_2, crs=states.crs, source=basemap)

for row in states.itertuples():
    x, y = row.centroid.x, row.centroid.y
    label = row.st_nm

    if label in ['Chandigarh']:
        x += 1 * 10**5
        
    ax_2.annotate(label, xy=(x,y), fontsize=5, ha='center')

plt.title('India: States and Union Territories (EPSG: 7755)')
plt.xlabel('Easting (m)')
plt.ylabel('Northing (m)')

plt.tight_layout()
plt.show()
```
