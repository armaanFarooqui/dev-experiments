```python
import fiona
import geopandas as gpd
import matplotlib.pyplot as plt
import colorcet as cc
```

```python
!ls ../data/
```

```python
fiona.listlayers('../data/india_admin_areas.geojson')
```

```python
states = gpd.read_file('../data/india_admin_areas.geojson')
states.shape
```

```python
states = states.dissolve(by='st_nm').reset_index()
states.shape
```

```python
print(states.crs)
states = states.to_crs('EPSG:7755')

print(states.crs)
```

```python
cmaps = ['Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 'Set1',
        'Set2', 'Set3', 'tab10', 'tab20', 'tab20b', 'tab20c']

fig_1, axes_1 = plt.subplots(nrows=4, ncols= 3, figsize=(12,12))
ax_1 = axes_1.flatten()

for ax, cmap in zip(ax_1, cmaps):
    states.plot(ax=ax, cmap=cmap, column='st_nm')
    ax.set_title(cmap)

plt.tight_layout()
plt.show()
```

```python
fig_2, ax_2 = plt.subplots(figsize=(10,8))
states.plot(ax=ax_2, column='st_nm', cmap='tab20b')

ax_2.set_title(f"India: States and Union Territories ({states.crs})")
ax_2.set_xlabel('Latitude')
ax_2.set_ylabel('Longitude')

plt.tight_layout()
plt.show()
```

```python
fig_3, ax_3 = plt.subplots(figsize=(10,8))
states.plot(ax=ax_3, column='st_nm', cmap=cc.cm.glasbey_cool_r)

for row in states.itertuples():
    x, y = row.geometry.centroid.x, row.geometry.centroid.y
    label = row.st_nm

    if label in ['Chandigarh', 'Jharkhand']:
        y-= 40000
        x-= 40000
        
    ax_3.annotate(label, xy=(x,y), fontsize=6, ha='center')

plt.tight_layout()
plt.show()
```
